```python
#IMPORT DEPENDENCY
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter

import warnings
warnings.filterwarnings('ignore')

from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced
```

```python
#LOAD SOCCER DATA INTO A DATAFRAME
soccer_data_load = "C:/Users/davec/Desktop/Class/20_Final_Project/soccer_data_V1.csv"
soccer_data_df = pd.read_csv(soccer_data_load, encoding = 'unicode_escape')
soccer_data_df.head(10)
```

```python
# Drop the null columns where all values are null
soccer_data_null_columns_df = soccer_data_df.dropna(axis='columns', how='all')
soccer_data_null_columns_df
```

```python
# Drop the null rows
df_null = soccer_data_df.dropna()
df_null

# No columns or rows were dropped so we can use our original data.
```

```python
soccer_data_df['Elite'] = np.where(soccer_data_df['points_per_game']> 2.25, "Elite", "Non-Elite")
soccer_data_df[['team_name','season','Elite']].head(200)
```

```python
past_soccer_df = soccer_data_df[(soccer_data_df["season"] == "2010/2011") | (soccer_data_df["season"] == "2011/2012") | (soccer_data_df["season"] == "2012/2013") | (soccer_data_df["season"] == "2013/2014") | (soccer_data_df["season"] == "2014/2015") | (soccer_data_df["season"] == "2015/2016") | (soccer_data_df["season"] == "2016/2017") | (soccer_data_df["season"] == "2017/2018")]
past_soccer_df
               
```

```python
current_soccer_df = soccer_data_df[(soccer_data_df["season"] == "2018/2019") | (soccer_data_df["season"] == "2019/2020") | (soccer_data_df["season"] == "2020/2021")]
current_soccer_df
```

```python
# Create our features
X = past_soccer_df.drop("Elite",axis=1)
X = pd.get_dummies(X)

# Create our target
y = current_soccer_df["Elite"]
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
rf = BalancedRandomForestClassifier(n_estimators = 100, random_state=1)
rf = rf.fit(X_train, y_train)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
y_pred = rf.predict(X_test)
balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```
