```python
#IMPORT DEPENDENCY
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter

import warnings
warnings.filterwarnings('ignore')

from sklearn import tree
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced
```

```python
#LOAD SOCCER DATA INTO A DATAFRAME
soccer_data_load = "C:/Users/davec/Desktop/Class/20_Final_Project/soccer_data_V2.csv"
soccer_data_df = pd.read_csv(soccer_data_load, encoding = 'unicode_escape')
soccer_data_df.head(10)
```

```python
# Drop the null columns where all values are null
soccer_data_null_columns_df = soccer_data_df.dropna(axis='columns', how='all')
soccer_data_null_columns_df
```

```python
# Drop the null rows
df_null = soccer_data_df.dropna()
df_null

# No columns or rows were dropped so we can use our original data.
```

```python
soccer_data_df['Champion'] = np.where(soccer_data_df['league_position'] == 1, "Champion", "Not")
soccer_data_df[['team_code','season','Champion']].head(200)
```

```python
soccer_data_df = soccer_data_df.drop(['team_name','common_name','country'], axis=1)
```

```python
soccer_data_df
```

```python
soccer_data_df['season'].dtypes
```

```python
past_soccer_df = soccer_data_df[(soccer_data_df["season"] == 2010) | (soccer_data_df["season"] == 2011) | (soccer_data_df["season"] == 2012) | (soccer_data_df["season"] == 2013) | (soccer_data_df["season"] == 2014) | (soccer_data_df["season"] == 2015) | (soccer_data_df["season"] == 2016) | (soccer_data_df["season"] == 2017)]
past_soccer_df     
```

```python
current_soccer_df = soccer_data_df[(soccer_data_df["season"] == 2018) | (soccer_data_df["season"] == 2019) | (soccer_data_df["season"] == 2020)]
current_soccer_df
```

### Split the Data into Training and Testing

```python
# Create our features
X_train = past_soccer_df.drop("Champion",axis=1)
X_test = current_soccer_df.drop("Champion",axis=1)
# X = pd.get_dummies(X)

# Create our target
y_train = past_soccer_df["Champion"]
y_test = current_soccer_df["Champion"]
```

### Balanced Random Forest Classifier

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
rf = BalancedRandomForestClassifier(n_estimators = 100, random_state=1)
rf = rf.fit(X_train, y_train)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
y_pred = rf.predict(X_test)
balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=100, random_state=65) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
